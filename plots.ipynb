```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

csv_files = ["Attention.csv", 
            "dsmil.csv", 
             "CLAM.csv", 
             "AttriMIL.csv", 
             "Emb_max.csv", 
             "Emb_mean.csv", 
             "GatedAttention.csv",
             "TransMIL.csv", 
             "VarMIL.csv",
             "ACMIL.csv"
            ]
plt.figure(figsize=(10, 7))

for i, csv_file in enumerate(csv_files):
    data = pd.read_csv(csv_file)
    fpr = data['fpr']
    tpr = data['tpr']
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f" {csv_files[i].replace('.csv', '')} (AUC = {roc_auc:.2f})")

plt.plot([0, 1], [0, 1], 'k--', label="Random guess")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc="lower right")
plt.savefig('roc_curves.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import auc


import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import auc

# Liste des fichiers CSV
csv_files = [
    "Attention.csv", 
    "dsmil.csv", 
    "CLAM.csv", 
    "AttriMIL.csv", 
    "Emb_max.csv", 
    "Emb_mean.csv", 
    "GatedAttention.csv",
    "ACMIL.csv"
]

# Liste des datasets
datasets = ['Fox', 'Elephant', 'Tiger', 'musk1', 'musk2']

# Crez une figure avec 5 sous-graphiques (1 ligne, 5 colonnes)
fig, axs = plt.subplots(2, 3, figsize=(18, 10), sharey=True)

axs = axs.flat


# Itrez sur les datasets et tracez les courbes ROC
for i, dataset in enumerate(datasets):
    for csv_file in csv_files:
        # Chargement des donnes
        file_path = f'csv_files/{dataset}/{csv_file}'
        data = pd.read_csv(file_path)
        
        # Calcul des courbes ROC et AUC
        fpr = data['fpr']
        tpr = data['tpr']
        roc_auc = auc(fpr, tpr)
        
        # Trace la courbe ROC pour chaque modle
        axs[i].plot(fpr, tpr, label=f"{csv_file.replace('.csv', '')} (AUC = {roc_auc:.2f})")
    
    # Ajoutez une ligne de rfrence pour le hasard
    axs[i].plot([0, 1], [0, 1], 'k--', label="Random guess")
    axs[i].set_xlim([0.0, 1.0])
    axs[i].set_ylim([0.0, 1.05])
    axs[i].set_xlabel('False Positive Rate')
    axs[i].set_title(f"ROC Curve - {dataset}")
    axs[i].legend(loc="lower right", fontsize=8)

fig.delaxes(axs[-1])

# Ajouter un label commun pour l'axe Y
fig.text(0.04, 0.5, 'True Positive Rate', va='center', rotation='vertical', fontsize=12)

# Titre global pour la figure
plt.suptitle('ROC Curves for Different Models and Datasets', fontsize=16)

# Ajuster les marges et sauvegarder
plt.tight_layout(rect=[0.03, 0.03, 1, 0.95])
plt.savefig('roc_curves_subplots.png')
plt.show()
```
